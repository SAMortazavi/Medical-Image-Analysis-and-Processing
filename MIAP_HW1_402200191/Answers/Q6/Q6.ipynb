```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
BrainCT_Image = cv.imread('brain_ct.jpeg', 0)
threshold_value = 80
thresholded_image = np.where(BrainCT_Image > threshold_value, 255, 0).astype(np.uint8)

kernel = np.ones((50,50), np.uint8)
kernel_2=np.ones((10,10), np.uint8)
dilated_image = cv.dilate(thresholded_image, kernel, iterations=1)
dilated_eroded_image = cv.erode(dilated_image, kernel_2, iterations = 7) 
plt.figure(figsize=(10,10)) 
plt.subplot(1,2,1)
plt.imshow(BrainCT_Image,cmap='gray')
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(dilated_eroded_image, cmap='gray')
plt.title("Masked Image")
plt.show()
```

```python
Noisy_Image=BrainCT_Image+np.random.normal(loc=0, scale=10, size=BrainCT_Image.shape)
threshold_value = 80
thresholded_image = np.where(Noisy_Image > threshold_value, 255, 0).astype(np.uint8)

kernel = np.ones((50,50), np.uint8)
kernel_2=np.ones((10,10), np.uint8)
dilated_image = cv.dilate(thresholded_image, kernel, iterations=1)
dilated_eroded_image = cv.erode(dilated_image, kernel_2, iterations = 7) 
plt.figure(figsize=(10,10)) 
plt.subplot(1,2,1)
plt.imshow(Noisy_Image,cmap='gray')
plt.title("Noisy Image")
plt.subplot(1,2,2)
plt.imshow(dilated_eroded_image, cmap='gray')
plt.title("Masked Image")
plt.show()
```
