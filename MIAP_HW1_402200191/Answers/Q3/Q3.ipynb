```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
Hand_img = cv.imread("hand_xray.jpg",0)
Hand_img_padded=np.zeros([900,900])
Hand_img_padded[:,50:850]=Hand_img
Hand_img_padded_fft= np.conj(np.fft.fft2(Hand_img_padded))
Hand_img_Rotated = np.fft.ifft2(Hand_img_padded_fft).real
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(Hand_img,cmap='gray')
plt.title("Real Image")
plt.subplot(1,2,2)
plt.imshow(Hand_img_Rotated[:,50:850],cmap='gray')
plt.title("Rotated Image")
plt.show()
```
