```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
lion_img=cv.imread('lion.jpg',0)
monkey_img=cv.imread('monkey.jpg',0)

monkey_fft=np.fft.fftshift(np.fft.fft2(monkey_img))
lion_fft=np.fft.fftshift(np.fft.fft2(lion_img))


lion_fft_amplitude = np.abs(lion_fft)
lion_fft_phase = np.angle(lion_fft)


monkey_fft_amplitude = np.abs(monkey_fft)
monkey_fft_phase = np.angle(monkey_fft)


Monkey_Lionphase = np.multiply(monkey_fft_amplitude, np.exp(1j * lion_fft_phase))
Monkey_Lionphase = np.real(np.fft.ifft2(Monkey_Lionphase))

Lion_Monkeyphase=np.multiply(lion_fft_amplitude, np.exp(1j * monkey_fft_phase))
Lion_Monkeyphase= np.real(np.fft.ifft2(Lion_Monkeyphase))

plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
plt.imshow(monkey_img,cmap='gray')
plt.title('Real monkey Image')
plt.subplot(2,2,2)
plt.imshow(lion_img,cmap='gray')
plt.title('Real lion Image')
plt.subplot(2,2,3)
plt.imshow(abs(Monkey_Lionphase),cmap='gray')
plt.title('Monkey amplitude & Lion phase')
plt.subplot(2,2,4)
plt.imshow(abs(Lion_Monkeyphase),cmap='gray')
plt.title('Lion amplitude & Monkey phase')
plt.show()
```
