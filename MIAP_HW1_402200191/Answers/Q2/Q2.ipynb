```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

In next section I've wrote the function's which is used to do the transforms.

```python
def transform(img1,transformname,gamma_list=[1]):
    img=img1.copy()
    if transformname=='PL':
        gamma_shape=np.array(gamma_list).shape
        plt.figure(figsize=(20,10))
        for i  in range(gamma_shape[1]): 
            gamma_corrected = np.array(255*(img / 255) ** gamma_list[0,i], dtype = np.uint8) 
            plt.subplot(int(gamma_shape[1]/5),5,i+1)
            plt.imshow(gamma_corrected,cmap='gray')
            plt.title(f'gamma={gamma_list[0,i]}')
        plt.show()
    elif transformname=="Log":
        c = 255/(np.log(1 + np.max(img))) 
        log_transformed = c * np.log(1 + img) 
        log_transformed = np.array(log_transformed, dtype = np.uint8) 
        plt.figure()
        plt.imshow(log_transformed,cmap='gray')
        plt.title("Log transformation")
        plt.show()
    return img
```

I have done the PL transform with different gamma values.

```python
Retina_img=cv.imread("retina.png",0)
gamma=np.linspace(0.1,1,10).reshape(1,10)
plt.imshow(Retina_img,cmap='gray')
plt.title('Original Image')
plt.show()
transform(Retina_img,'PL',gamma)
transform(Retina_img,'Log');
```

```python
Retina_img=cv.imread("retina.png",0)
gamma=np.linspace(1,10,10).reshape(1,10)
transform(Retina_img,'PL',gamma)
plt.show()
```
